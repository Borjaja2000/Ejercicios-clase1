Crear una clase Círculo que tenga un atributo radio y dos métodos para calcular su área y su perímetro.

```python
from math import pi
class Circulo:
    def __init__(self, radio):
        self.radio = radio
    def calcular_area(self):
        area = pi * self.radio**2
        return area
    def calcular_perimetro(self):
        perimetro = 2*pi*self.radio
        return perimetro

circulo = Circulo(10)
print(f'El area del circulo es: {circulo.calcular_area()} metros cuadrados')
print(f'El perímetro del circulo es: {circulo.calcular_perimetro()} metros')
```

Crear una clase Rectángulo con atributos largo y ancho. La clase debe tener métodos para calcular su área y su perímetro.

```python
class Rectangulo:
    def __init__(self, largo, ancho):
        self.largo = largo
        self.ancho = ancho
    def calcular_area(self):
        area = self.largo * self.ancho
        return area
    def calcular_perimetro(self):
        perimetro = 2*self.largo + 2*self.ancho
        return perimetro

rectangulo = Rectangulo(10, 20)
print(f'El area del rectángulo es: {rectangulo.calcular_area()} metros cuadrados')
print(f'El perímetro del rectángulo es: {rectangulo.calcular_perimetro()} metros')
```

Crear una clase CuentaBancaria que tenga atributos titular y saldo. La clase debe tener métodos para depositar, retirar y mostrar el saldo actual.



```python
class CuentaBancaria:
    def __init__(self, titular, saldo):
        self.titular = titular
        self.saldo = saldo
    def depositar(self, dinero):
        if(dinero>0):
            self.saldo +=dinero
            return f"Ha depositado {dinero}€"
        else:
            return "Introduce una cantidad correcta"

    def retirar(self, dinero):
        if(0<dinero<=self.saldo):
            self.saldo-=dinero
        else:
            return "Introduce una cantidad correcta"
        return f"Ha retirado {dinero}€"
    def mostrar_saldo(self):
        return f"Su saldo actual es {self.saldo}€"

cuenta = CuentaBancaria("Borja",0)
print(cuenta.depositar(20))
print(cuenta.retirar(1))
print(cuenta.mostrar_saldo())
```

Crear una clase Estudiante con atributos nombre, edad, grado y calificaciones (un diccionario con nombres de materias y calificaciones). La clase debe tener métodos para agregar una calificación, calcular el promedio de las calificaciones y mostrar el promedio.



```python
class Estudiante:
    def __init__(self, nombre,edad,grado,calificaciones):
        self.nombre = nombre
        self.edad = edad
        self.grado = grado
        self.calificaciones = calificaciones
    def agregar_calificacion(self, materia, calificacion):
        self.calificaciones[materia] = calificacion
        return self.calificaciones
    def calcular_promedio(self):
        lista = list(self.calificaciones.values())
        promedio = sum(lista) / len(lista)
        return promedio
    def mostrar_promedio(self):
        return f"El promedio de {self.nombre} es {self.calcular_promedio()}" 
    
estudiante = Estudiante("Borja",24,"Curso UpgradeHub",{})
print(estudiante.agregar_calificacion("Matematicas",10))
print(estudiante.agregar_calificacion("Lengua",9))
estudiante.calcular_promedio()
print(estudiante.mostrar_promedio())
```

Crear una jerarquía de clases para representar figuras geométricas. Incluir clases para Figura, Círculo, Rectángulo y Triángulo. La clase Figura debe ser una superclase abstracta con métodos para calcular el área y el perímetro. Las subclases deben implementar estos métodos.



```python
import math
class Figura:
    def __init__(self):
        pass
    def area(self):
        pass
    def perimetro(self):
        pass
class Circulo(Figura):
    def __init__(self, radio):
        self.radio = radio
    def area(self):
        return math.pi*self.radio**2
    def perimetro(self):
        return 2*pi*self.radio
class Rectangulo(Figura):
    def __init__(self, largo, ancho):
        self.largo = largo
        self.ancho = ancho
    def area(self):
        return self.largo*self.ancho
    def perimetro(self):
        return 2*(self.largo + self.ancho)
class Triangulo(Figura):
    def __init__(self, lado1, lado2, lado3):
        self.lado1 = lado1
        self.lado2 = lado2
        self.lado3 = lado3
    def area(self):
        s = (self.lado1 + self.lado2 + self.lado3) / 2
        return math.sqrt(s * (s - self.lado1) * (s - self.lado2) * (s - self.lado3))
    def perimetro(self):
        return self.lado1+self.lado2+self.lado3
    
circulo = Circulo(5)
print("Área del círculo:", circulo.area())
print("Perímetro del círculo:", circulo.perimetro())

rectangulo = Rectangulo(4, 6)
print("Área del rectángulo:", rectangulo.area())
print("Perímetro del rectángulo:", rectangulo.perimetro())

triangulo = Triangulo(3, 4, 5)
print("Área del triángulo:", triangulo.area())
print("Perímetro del triángulo:", triangulo.perimetro())
```

Crear una clase Automóvil con atributos marca, modelo, año y kilometraje. Crear una clase Concesionario que contenga una lista de objetos Automóvil. La clase Concesionario debe tener métodos para agregar un automóvil, eliminar un automóvil y buscar automóviles por marca, modelo o año.



```python
class Automovil:
    def __init__(self, marca, modelo, año, kilometraje):
        self.marca = marca
        self.modelo = modelo
        self.año = año
        self.kilometraje = kilometraje

    def __str__(self):
        return f"{self.marca} {self.modelo} - Año: {self.año} - Kilometraje: {self.kilometraje}"


class Concesionario:
    def __init__(self):
        self.inventario = []

    def agregar_automovil(self, automovil):
        self.inventario.append(automovil)
        print("Automóvil agregado al inventario.")

    def eliminar_automovil(self, marca, modelo):
        for automovil in self.inventario:
            if automovil.marca.lower() == marca.lower() and automovil.modelo.lower() == modelo.lower():
                self.inventario.remove(automovil)
                print("Automóvil eliminado del inventario.")
                return
        print("El automóvil no se encontró en el inventario.")

    def buscar_automovil_por_marca(self, marca):
        encontrados = [auto for auto in self.inventario if auto.marca.lower() == marca.lower()]
        if encontrados:
            for auto in encontrados:
                print(auto)
        else:
            print("No se encontraron automóviles de esa marca en el inventario.")

    def buscar_automovil_por_modelo(self, modelo):
        encontrados = [auto for auto in self.inventario if auto.modelo.lower() == modelo.lower()]
        if encontrados:
            for auto in encontrados:
                print(auto)
        else:
            print("No se encontraron automóviles de ese modelo en el inventario.")

    def buscar_automovil_por_año(self, año):
        encontrados = [auto for auto in self.inventario if auto.año == año]
        if encontrados:
            for auto in encontrados:
                print(auto)
        else:
            print("No se encontraron automóviles de ese año en el inventario.")


# Ejemplo de uso:
if __name__ == "__main__":
    # Crear concesionario
    concesionario = Concesionario()

    # Agregar automóviles al inventario
    auto1 = Automovil("Toyota", "Corolla", 2018, 50000)
    auto2 = Automovil("Honda", "Civic", 2020, 30000)
    auto3 = Automovil("Ford", "Fiesta", 2019, 40000)
    concesionario.agregar_automovil(auto1)
    concesionario.agregar_automovil(auto2)
    concesionario.agregar_automovil(auto3)

    # Buscar automóviles por marca
    print("\nBuscando automóviles por marca Toyota:")
    concesionario.buscar_automovil_por_marca("Toyota")

    # Buscar automóviles por modelo
    print("\nBuscando automóviles por modelo Civic:")
    concesionario.buscar_automovil_por_modelo("Civic")

    # Buscar automóviles por año
    print("\nBuscando automóviles del año 2019:")
    concesionario.buscar_automovil_por_año(2019)

    # Eliminar un automóvil
    print("\nEliminando automóvil Honda Civic del inventario:")
    concesionario.eliminar_automovil("Honda", "Civic")

    # Mostrar inventario después de eliminar un automóvil
    print("\nInventario después de eliminar el automóvil:")
    for auto in concesionario.inventario:
        print(auto)
```

Crear una jerarquía de clases para representar empleados en una empresa. Incluir clases para Empleado, Gerente, Ingeniero y Vendedor. La clase Empleado debe ser una superclase con atributos comunes, como nombre, edad y salario. Las subclases deben tener atributos y métodos específicos, como bonificaciones para el Gerente, área de especialización para el Ingeniero y comisiones para el Vendedor.



```python
class Empleado:
    def __init__(self , nombre, edad, salario):
        self.nombre = nombre
        self.edad = edad
        self.salario = salario
    def __str__(self):
        return f"Nombre: {self.nombre}, Edad: {self.edad}, Salario: {self.salario}"
class Gerente(Empleado):
    def __init__(self, nombre, edad, salario, bonificacion):
        super().__init__(nombre, edad, salario)
        self.bonificacion = bonificacion
    def __str__(self):
        return super().__str__() + f", Bonificación: {self.bonificacion}"
class Ingeniero(Empleado):
    def __init__(self, nombre, edad, salario, especializacion):
        super().__init__(nombre, edad, salario)
        self.especializacion = especializacion
    def __str__(self):
        return super().__str__() + f", Especialización: {self.especializacion}"

class Vendedor(Empleado):
    def __init__(self, nombre, edad, salario, comision):
        super().__init__(nombre, edad, salario)
        self.comision = comision
    def __str__(self):
        return super().__str__() + f", Comisión: {self.comision}"


# Ejemplo de uso
gerente = Gerente("Juan", 40, 60000, 10000)
ingeniero = Ingeniero("María", 35, 55000, "Ingeniería de Software")
vendedor = Vendedor("Carlos", 25, 45000, 5000)

print(gerente)
print(ingeniero)
print(vendedor)
```

Crear un sistema de gestión de bibliotecas utilizando clases. Crear clases para Libro, Autor, Biblioteca y Usuario. Los libros deben tener atributos como título, autor y disponibilidad. La biblioteca debe tener una lista de libros y métodos para agregar libros, eliminar libros, buscar libros por título o autor y prestar libros a usuarios. Los usuarios deben tener una lista de libros prestados y métodos para devolver libros.



```python
class Autor:
    def __init__(self, nombre, apellido):
        self.nombre = nombre
        self.apellido = apellido
    def __str__(self):
        return f"{self.nombre} {self.apellido}"
    
class Libro:
    def __init__(self, titulo, autor):
        self.titulo = titulo
        self.autor = autor
        self.disponible = True
    def __str__(self):
        return f'Título: {self.titulo}, Autor: {self.autor}, Disponibilidad: {self.disponible}'

class Biblioteca():
    def __init__(self):
        self.libros = []
    def agregar_libro(self, libro):
        self.libros.append(libro)
    def eliminar_libro(self, libro):
        if libro in self.libros:
            self.libros.remove(libro)
            print("Libro elimando de la biblioteca")
        else:
            print("No existe libro en la biblioteca")
                
    def buscar_libro_por_titulo(self, titulo):
         for libro in self.libros:
            if libro.titulo == titulo:
                print(libro) 
            else:
                print("No existe ningun libro con ese título")
    def buscar_libro_por_autor(self, autor):
         for libro in self.libros:
            if libro.autor.nombre == autor:
                print(libro)
            else:
                print("No existe ningun libro con ese autor")
    def prestar_libro(self, libro, usuario):
        if libro in self.libros:
            if libro.disponible:
                libro.disponible = False
                usuario.agregar_libro_prestado(libro)
                print("Libro prestado correctamente")
            else:
                print("El libro no esta disponible")
    def mostrar_libros(self):
        if self.libros:
            for libro in self.libros:
                print(libro)
        else:
            print("La biblioteca está vacía.")

class Usuario():
    def __init__(self, nombre):
        self.nombre = nombre
        self.libros_prestados = []

    def agregar_libro_prestado(self, libro):
        self.libros_prestados.append(libro)

    def devolver_libro(self, libro):
        if libro in self.libros_prestados:
            self.libros_prestados.remove(libro)
            libro.disponible = True
            print("Libro devuelto correctamente")
        else:
            print("No tiene este libro prestado")




# Crear autores
autor1 = Autor("Gabriel", "García Márquez")
autor2 = Autor("J.K.", "Rowling")

# Crear libros
libro1 = Libro("Cien años de soledad", autor1)
libro2 = Libro("Harry Potter y la piedra filosofal", autor2)

# Crear biblioteca
biblioteca = Biblioteca()

# Agregar libros a la biblioteca
biblioteca.agregar_libro(libro1)
biblioteca.agregar_libro(libro2)

# Mostrar libros en la biblioteca
print("Libros en la biblioteca:")
biblioteca.mostrar_libros()

# Buscar libro por título
print("\nBuscando libro por título:")
biblioteca.buscar_libro_por_titulo("cien años de soledad")

# Buscar libro por autor
print("\nBuscando libro por autor:")
biblioteca.buscar_libro_por_autor("Gabriel")

# Crear usuario
usuario = Usuario("Juan")

# Prestar libro a usuario
biblioteca.prestar_libro(libro1, usuario)

# Mostrar libros prestados por el usuario
print(f"\nLibros prestados a {usuario.nombre}:")
for libro in usuario.libros_prestados:
    print(libro)

# Devolver libro
usuario.devolver_libro(libro1)

# Mostrar libros en la biblioteca después de la devolución
print("\nLibros en la biblioteca después de la devolución:")
biblioteca.mostrar_libros()
```
