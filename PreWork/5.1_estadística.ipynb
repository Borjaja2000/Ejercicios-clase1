# Estadística descriptiva

## Conceptos básicos de estadística descriptiva
La estadística descriptiva es una rama de la estadística que se enfoca en resumir, describir y analizar un conjunto de datos. En el campo del análisis de datos, la estadística descriptiva ayuda a los analistas a comprender las características y patrones generales de los datos recopilados, lo que les permite tomar decisiones informadas y construir modelos predictivos.

Las técnicas de estadística descriptiva incluyen medidas de tendencia central, medidas de dispersión y visualización de datos. Algunos ejemplos de aplicaciones de la estadística descriptiva en el análisis de datos son:

- Entender el comportamiento del cliente en una tienda en línea.
- Analizar las ventas de productos en función de variables como ubicación, género y edad.
- Evaluar el rendimiento de los estudiantes en una clase.
- Comparar diferentes acciones en el mercado financiero.
- Identificar patrones en datos de sensores en un sistema de IoT.

La estadística descriptiva es esencial para los analistas de datos, ya que proporciona una base sólida para la exploración de datos, la detección de patrones y tendencias, y la identificación de variables clave en un conjunto de datos. Además, la estadística descriptiva es el primer paso en el proceso de análisis de datos antes de realizar análisis inferenciales y predictivos.

En la siguiente sección, presentaremos medidas de tendencia central y medidas de dispersión, que son componentes clave de la estadística descriptiva.


## Medidas de tendencia central: media, mediana, moda

Las medidas de tendencia central son valores que describen la posición central de un conjunto de datos. Ayudan a resumir un conjunto de datos con un solo valor que representa el "centro" de la distribución. Las tres medidas más comunes de tendencia central son la media, la mediana y la moda.

### Media
La media, también conocida como el promedio, es la suma de todos los valores en un conjunto de datos dividida por el número total de valores. La media es sensible a valores extremos o atípicos, lo que significa que puede verse afectada significativamente por estos valores.

```python


````python
def media(datos):
    return sum(datos) / len(datos)
````

### Mediana
La mediana es el valor que separa un conjunto de datos en dos mitades, de modo que la mitad de los valores son menores o iguales que la mediana y la otra mitad son mayores o iguales. Para encontrar la mediana, primero se deben ordenar los datos. Si el número de datos es impar, la mediana es el valor del centro. Si el número de datos es par, la mediana es el promedio de los dos valores centrales. A diferencia de la media, la mediana es menos sensible a valores extremos.

```python


````python
def mediana(datos):
    datos_ordenados = sorted(datos)
    n = len(datos)
    if n % 2 == 0:
        return (datos_ordenados[n // 2 - 1] + datos_ordenados[n // 2]) / 2
    else:
        return datos_ordenados[n // 2]
````

### Moda
La moda es el valor que ocurre con mayor frecuencia en un conjunto de datos. Puede haber más de una moda en un conjunto de datos (multimodal) si varios valores ocurren con la misma frecuencia máxima. La moda es útil para identificar el valor más común en un conjunto de datos, pero no siempre es representativa de la posición central del conjunto de datos.

```python


````python
def moda(datos):
    frecuencia = {}
    for d in datos:
        if d in frecuencia:
            frecuencia[d] += 1
        else:
            frecuencia[d] = 1
    return max(frecuencia, key=frecuencia.get)
````

## Ejemplo:

````python
datos = [3, 5, 8, 3, 9, 5, 2, 8]
media_datos = media(datos)
mediana_datos = mediana(datos)
moda_datos = moda(datos)

print(f"Media: {media_datos}\nMediana: {mediana_datos}\nModa: {moda_datos}")

"""
En este ejemplo, la media es 5.375, lo que indica que el promedio de todos los valores en el conjunto de datos es 5.375.
La mediana es 5.0, lo que significa que el valor del centro del conjunto de datos es 5 cuando los datos están ordenados. 
La moda es 3, 5 y 8, ya que estos valores ocurren con la misma frecuencia máxima de 2 veces cada uno. """
````

## Medidas de dispersión: desviación estándar, varianza

Las medidas de dispersión son valores que describen la variabilidad o dispersión en un conjunto de datos. Estas medidas nos ayudan a entender qué tan dispersos están los datos alrededor de la medida de tendencia central, como la media. Las dos medidas de dispersión más comunes son la varianza y la desviación estándar.

### Varianza
La varianza es una medida que representa la dispersión de los valores en un conjunto de datos en relación con la media. La varianza se calcula como el promedio de las diferencias al cuadrado entre cada valor en el conjunto de datos y la media. Una varianza alta indica que los valores en el conjunto de datos están muy dispersos, mientras que una varianza baja indica que los valores están más agrupados alrededor de la media.



````python
def varianza(datos):
    m = media(datos)
    return sum([(d - m)**2 for d in datos]) / len(datos)
````

### Desviación estándar
La desviación estándar es la raíz cuadrada de la varianza y se utiliza comúnmente como medida de dispersión en la estadística descriptiva. La desviación estándar tiene las mismas unidades que los datos, lo que facilita la interpretación de la dispersión de los valores. Una desviación estándar alta indica que los valores en el conjunto de datos están muy dispersos, mientras que una desviación estándar baja indica que los valores están más agrupados alrededor de la media.

````python
def desviacion_estandar(datos):
    return varianza(datos)**0.5
````

Estas medidas de dispersión son importantes en el análisis de datos, ya que proporcionan información sobre la variabilidad y la dispersión de un conjunto de datos. Junto con las medidas de tendencia central, las medidas de dispersión nos ayudan a obtener una comprensión más completa de la distribución de los datos y a identificar patrones y tendencias.

### Ejemplos:


````python

datos = [12, 15, 21, 15, 27, 30, 18, 22]
varianza_datos = varianza(datos)
desviacion_datos = desviacion_estandar(datos)

print(f"Varianza: {varianza_datos}\nDesviación estándar: {desviacion_datos}")
````

## Distribuciones de probabilidad: normal, uniforme, binomial, Poisson

Las distribuciones de probabilidad son modelos matemáticos que describen cómo se distribuyen las variables aleatorias en un conjunto de datos. Las distribuciones de probabilidad son fundamentales en el análisis de datos, ya que nos ayudan a comprender la variabilidad y la incertidumbre en los datos y a realizar inferencias estadísticas. A continuación, se presentan algunas de las distribuciones de probabilidad más comunes:





### Distribución normal
La distribución normal, también conocida como distribución gaussiana, es una distribución de probabilidad continua que tiene una forma de campana. La distribución normal está definida por dos parámetros: la media (µ) y la desviación estándar (σ). La media indica la posición central de la distribución, mientras que la desviación estándar indica la dispersión de los valores alrededor de la media.

La distribución normal es ampliamente utilizada en el análisis de datos debido a su propiedad de que muchos fenómenos naturales y sociales siguen una distribución normal, como las calificaciones de los estudiantes, las alturas de las personas o los errores de medición.



Algunos ejemplos de problemas que nos podemos encontrar con este tipo de distribución: 

Análisis de ventas: Eres un analista de datos en una empresa de ventas y estás interesado en analizar las ventas diarias de un producto específico. Observas que las ventas siguen una distribución normal con una media de 500 unidades y una desviación estándar de 50 unidades. Puedes utilizar la distribución normal para estimar la probabilidad de vender una cantidad específica de unidades en un día dado o determinar qué días tuvieron ventas excepcionalmente altas o bajas.

Control de calidad: Imagina que trabajas como analista de datos en una fábrica que produce tornillos. La longitud de los tornillos producidos sigue una distribución normal con una media de 10 mm y una desviación estándar de 0.1 mm. Puedes utilizar la distribución normal para determinar qué proporción de tornillos está dentro de los límites de tolerancia aceptables o identificar tornillos defectuosos que estén fuera de estos límites.

Análisis de tiempo de espera: Supón que eres un analista de datos en una empresa de transporte y estás interesado en analizar los tiempos de espera de los autobuses en una parada específica. Los tiempos de espera siguen una distribución normal con una media de 10 minutos y una desviación estándar de 2 minutos. Puedes utilizar la distribución normal para estimar la probabilidad de que un autobús llegue dentro de un cierto tiempo o determinar si un autobús en particular tiene un tiempo de espera significativamente diferente al promedio.

Análisis de opiniones de clientes: Imagina que trabajas como analista de datos en una empresa que recopila opiniones de clientes sobre productos y servicios. Las puntuaciones de satisfacción de los clientes siguen una distribución normal con una media de 4 estrellas (de 5) y una desviación estándar de 0.5 estrellas. Puedes utilizar la distribución normal para determinar qué productos o servicios tienen puntuaciones de satisfacción significativamente diferentes a la media o identificar áreas en las que se necesita mejorar.

````python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros de la distribución normal
media = 100
desviacion_estandar = 15

# Genera puntos de datos para el eje x
x = np.linspace(media - 3 * desviacion_estandar, media + 3 * desviacion_estandar, 100)

# Función para calcular la función de densidad de probabilidad (PDF) de la distribución normal
def distribucion_normal(x, media, desviacion_estandar):
    exponente = -((x - media) ** 2) / (2 * (desviacion_estandar ** 2))
    return (1 / (desviacion_estandar * np.sqrt(2 * np.pi))) * np.exp(exponente)

# Calcula los valores de la función de densidad de probabilidad para cada punto en x
y = distribucion_normal(x, media, desviacion_estandar)

# Gráfica la distribución normal
plt.plot(x, y)
plt.title('Distribución normal con media = 100 y desviación estándar = 15')
plt.xlabel('Valor')
plt.ylabel('Densidad de probabilidad')
plt.grid()
plt.show()
````

### Distribución uniforme
La distribución uniforme es una distribución de probabilidad continua en la que todos los valores en un intervalo tienen la misma probabilidad de ocurrencia. La distribución uniforme está definida por dos parámetros: el límite inferior (a) y el límite superior (b). La distribución uniforme es útil para modelar fenómenos en los que cada resultado es igualmente probable, como lanzar un dado justo o seleccionar una carta al azar de un mazo.



Algunos ejemplos de problemas que nos podemos encontrar con este tipo de distribución: 


Simulación de tiempos de llegada: Supongamos que estás analizando el tiempo que tardan los clientes en llegar a un restaurante entre las 6 p.m. y las 10 p.m. Se sabe que las llegadas son uniformemente distribuidas en ese intervalo de tiempo. Podrías utilizar la distribución uniforme para calcular la probabilidad de que un cliente llegue durante un periodo específico o determinar cuántos clientes se esperan en total durante un intervalo de tiempo dado.

Análisis de inventario: Imagina que trabajas en una empresa que vende productos con una vida útil limitada, como alimentos perecederos. Se sabe que la vida útil de estos productos varía uniformemente entre 5 y 15 días. Puedes utilizar la distribución uniforme para estimar la probabilidad de que un producto se eche a perder dentro de un cierto número de días y planificar la rotación y reposición de inventario de manera eficiente.

Estudio de tiempos de respuesta: Supón que eres un analista de datos en una empresa de soporte técnico y estás interesado en analizar los tiempos de respuesta de los agentes de soporte. Los tiempos de respuesta de los agentes varían uniformemente entre 1 y 10 minutos. Puedes utilizar la distribución uniforme para calcular la probabilidad de que un agente responda dentro de un cierto tiempo y determinar si es necesario mejorar los tiempos de respuesta o la eficiencia del equipo de soporte.

Análisis de fallas de sistemas: Imagina que trabajas en una empresa que opera servidores y sistemas informáticos. Se sabe que la probabilidad de falla de un componente específico sigue una distribución uniforme entre 0 y 1000 horas de funcionamiento. Puedes utilizar la distribución uniforme para estimar la probabilidad de que un componente falle en un intervalo de tiempo específico y planificar el mantenimiento preventivo y la sustitución de componentes en consecuencia.

````python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros de la distribución uniforme
minimo = 20
maximo = 80

# Genera puntos de datos para el eje x
x = np.linspace(minimo, maximo, 100)

# Función para calcular la función de densidad de probabilidad (PDF) de la distribución uniforme
def distribucion_uniforme(x, minimo, maximo):
    pdf = np.zeros_like(x)
    pdf[(x >= minimo) & (x <= maximo)] = 1 / (maximo - minimo)
    return pdf

# Calcula los valores de la función de densidad de probabilidad para cada punto en x
y = distribucion_uniforme(x, minimo, maximo)

# Gráfica la distribución uniforme
plt.plot(x, y)
plt.title('Distribución uniforme entre 20 y 80')
plt.xlabel('Valor')
plt.ylabel('Densidad de probabilidad')
plt.grid()
plt.show()

````

### Distribución binomial
La distribución binomial es una distribución de probabilidad discreta que modela el número de éxitos en una secuencia de n ensayos independientes, cada uno con una probabilidad de éxito p. La distribución binomial está definida por dos parámetros: el número de ensayos (n) y la probabilidad de éxito (p). La distribución binomial es útil para modelar fenómenos como el número de caras en una serie de lanzamientos de monedas, la cantidad de aprobados en un examen o la cantidad de ventas exitosas en una campaña de marketing.



Problemas con este tipo de distribución que podemos encontrarnos: 

Análisis de conversión de sitios web: Supongamos que trabajas en una empresa de comercio electrónico y estás interesado en analizar la tasa de conversión de visitantes a compradores en un sitio web. Puedes utilizar la distribución binomial para determinar la probabilidad de que un número específico de visitantes hagan una compra y estimar la tasa de conversión con un cierto nivel de confianza.

Análisis de calidad de producción: Imagina que trabajas en una fábrica que produce componentes electrónicos y estás interesado en analizar la calidad de la producción. Se sabe que la probabilidad de que un componente tenga un defecto es del 2%. Puedes utilizar la distribución binomial para estimar el número de componentes defectuosos en un lote determinado y determinar si la producción cumple con los estándares de calidad.

Análisis de encuestas de opinión: Supón que eres un analista de datos en una empresa que realiza encuestas de opinión pública. Estás interesado en analizar la tasa de aprobación de un candidato político en una población específica. Puedes utilizar la distribución binomial para estimar la probabilidad de que un cierto número de personas aprueben al candidato y calcular un intervalo de confianza para la tasa de aprobación.

Análisis de éxito de marketing: Imagina que trabajas en una empresa de marketing y estás interesado en analizar el éxito de una campaña publicitaria. Se sabe que la tasa de respuesta a la campaña es del 5%. Puedes utilizar la distribución binomial para estimar el número de respuestas esperadas y determinar si la campaña fue exitosa o no.

````python
# Parámetros de la distribución binomial
n = 10
p = 0.5

# Función para calcular el coeficiente binomial
def coeficiente_binomial(n, k):
    if k == 0 or k == n:
        return 1
    return coeficiente_binomial(n - 1, k - 1) + coeficiente_binomial(n - 1, k)

# Función para calcular la función de masa de probabilidad (PMF) de la distribución binomial
def distribucion_binomial(n, p, k):
    q = 1 - p
    return coeficiente_binomial(n, k) * (p ** k) * (q ** (n - k))

# Genera puntos de datos para el eje x
x = list(range(n + 1))

# Calcula los valores de la función de masa de probabilidad para cada punto en x
y = [distribucion_binomial(n, p, k) for k in x]

# Gráfica la distribución binomial
plt.bar(x, y)
plt.title('Distribución binomial con n = 10 y p = 0.5')
plt.xlabel('Éxitos')
plt.ylabel('Probabilidad')
plt.grid()
plt.show()
````
